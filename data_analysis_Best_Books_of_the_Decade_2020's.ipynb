**Data Analysis on 'Best Books of the Decade 2020's**
**Problem/ask**
What books became popular with the highest ratings and number of votes of the 2020's decade

**Data Source**
https://www.kaggle.com/datasets/valakhorasani/best-books-of-the-decade-2020s

**Data Preparation and Analysis**
1. import dataset into google colab
2. unzip dataset
3. load and display datasets using the pandas library.
4. using the pandas library to find out dataset information.
5. perform cleaning checks on the dataset.
6. Change the data type in the Rating column from object data type to float.
7. fill in the NaN value in the Rating column with the average value.
8. Change the data type in the 'Number of Votes' column from the object data type to Integer.
9. check data duplication
10. visualize data

**Analysis Insights**
1. The 5 books with the highest number of ratings are books with the titles Anshuman Krit Saral Vastu Gyan, Metaphysics of Shound, The Water Walls, Anzhula: According to ISH (New Beginning M/M Series), In Her Shoes: Women of the Eighth: A Memoir and Anthology.
2. The book that has the most votes, reaching more than 50,000, is Light Bringer (Red Rising Saga, #6), but its rating is in 10th place compared to Anshuman Krit Saral Vastu Gyan's book which has the highest rating even though the number of votes is 0.
3. If you look at the score, the book The Invisible Life of Addie LaRue is the most popular book with a total score reaching above 60,000
4. The most popular author is Victoria Schwab with the book title The Invisible Life of Addie LaRue with a total score above 60,000

**Recomendations**
Most of the books that are popular and have lots of fans have low ratings and books that have a small number of fans have very good ratings. Recommendations that can be given based on the results of the analysis are that promoting books with high ratings can utilize online media such as social media and so on. Apart from that, the recommendation given is to improve the quality of books that have a high number of enthusiasts by increasing the interest of the book, its uniqueness and also a more interesting storyline.

```python
import pandas as pd

df=pd.read_csv("books_of_the_decade.csv")
df.head()
```

```python
df.info()
```

```python
# df['Rating']= df['Rating'].astype(float)
print(df['Rating'].dtype)
```

```python
# mengkonversi kolom 'rating' menjadi float dan menangani nilai yang tidak valid
df['Rating'] = pd.to_numeric(df['Rating'],errors='coerce')
```

```python
# memeriksa kembali tipe data dan nilai yang mungkin menjadi NaN
print(df['Rating'].dtype)
print(df['Rating'].isnull().sum()) #mengecek berapa banyak nilai NaN
```

```python
# mengisi nilai NaN dengan rata-rata rating
df['Rating'].fillna(df['Rating'].mean(),inplace=True)
```

```python
print(df['Rating'].isnull().sum())
```

```python
# memeriksa apakah kolom 'rating' sudah bersih dan sesuai
print(df['Rating'].head())
print(df.info())
```

```python
# mengecek type data 'Number of Votes'
print(df['Number of Votes'].dtype)
```

```python
df.head()
```

```python
# menghapus koma di kolom 'Number of Votes' sebelum cleaning
df['Number of Votes'] = df['Number of Votes'].str.replace(',','')
```

```python
df.head()
```

```python
# konversi kolom 'Number of Votes' menjadi integer
# errors='coerce' : akan mengganti nilai yang tidak dapat diubah menjadi angka dengan NaN
df['Number of Votes'] = pd.to_numeric(df['Number of Votes'], errors='coerce',downcast='integer')
```

```python
print(df['Number of Votes'].dtype)
print(df['Number of Votes'].isnull().sum())
```

```python
# mengisi NaN dengan 0 atau nilai lain sebelum dikonversi ke integer
df['Number of Votes'].fillna(0,inplace=True)
# konversi type data ke integer setelah NaN diisi
df['Number of Votes']=df['Number of Votes'].astype(int)
```

```python
print(df['Number of Votes'].dtype)
print(df['Number of Votes'].isnull().sum())
```

```python
df.head()
```

```python
# mengecek adanya data duplikat
duplicated_rows=df[df.duplicated()]

# menampilkan baris duplikat
print(duplicated_rows)
```

```python
# menghitung jumlah baris duplikat
num_duplicates=df.duplicated().sum()
print(f"Jumlah baris duplikat : {num_duplicates}")
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# mengurutkan data berdasarkan 'rating'
top_10_books = df.sort_values(by='Rating', ascending=False).head(5)

# membuat barplot untuk menampilkan 10 judul buku dengan rating terbanyak
plt.figure(figsize=(10,6))
sns.barplot(x='Rating', y='Book Name', data=top_10_books,palette='viridis')

# menmabahkan judul dan label sumbu
plt.title('5 Judul Buku dengan Jumlah Rating Terbanyak', fontsize=16)
plt.xlabel('Jumlah Rating', fontsize=12)
plt.ylabel('Judul Buku', fontsize=12)

# menampilkan plot
plt.show()
```

```python
#ambil 10 buku teratas berdasarkan jumlah reviews dan votes terbanyak
top_10_books=df.sort_values(by=['Rating','Number of Votes'], ascending=False).head(10)

# membuat plot dengan 2 sumbu y
fig,ax1 = plt.subplots(figsize=(10,6))

# plot pertama untuk jumlah rating
ax1.bar(top_10_books['Book Name'],top_10_books['Rating'],color='b',alpha=0.6,label='Rating')
ax1.set_xlabel('Judul Buku')
ax1.set_ylabel('Jumlah Reviews (Rating)', color='b')
ax1.tick_params(axis='y',labelcolor='b')
ax1.set_xticklabels(top_10_books['Book Name'],rotation=45,ha='right')

# plot kedua untuk jumlah votes
ax2=ax1.twinx() # membuat sumbu y kedua yang berbagi sumbu x
ax2.plot(top_10_books['Book Name'],top_10_books['Number of Votes'],color='r',marker='o',label='Number of Votes')
ax2.set_ylabel('Jumlah Votes',color='r')
ax2.tick_params(axis='y',labelcolor='r')

# menmabahkan judul
plt.title('Top 10 Judul Buku dengan Jumlah Rating dan Votes terbanyak')

# menampilkan plot
plt.show()
```

```python
# ambil 10 buku teratas berdasarkan score (rating) terbanyak
top_10_books_by_score = df.sort_values(by='Score', ascending=False).head(10)

# membuat scatter plot untuk menunjukkan popularitas buku berdasarkan score
plt.figure(figsize=(10,6))
sns.scatterplot(x='Score',y='Book Name', size='Score', hue='Score', sizes=(100,1000),
                data=top_10_books_by_score,palette='coolwarm',legend=False)

# menambahkan judul dan label
plt.title('Popularitas Buku Berdasarkan Score', fontsize=16)
plt.xlabel('Score', fontsize=12)
plt.xticks(rotation=45,ha='right')

# menampilkan plot
plt.show()
```

```python
# ambil 10 author berdasarkan score
top_10_author_by_score=df.sort_values(by='Score',ascending=False).head(10)

# membuat scatter plot untuk menunjukkan popularitas author berdasarkan score
plt.figure(figsize=(10,6))
sns.scatterplot(x='Score',y='Author', size='Score', hue='Score', sizes=(100,1000),
                data=top_10_author_by_score, palette='coolwarm', legend=False)

# membuat judul dan label
plt.title('Popularitas Author berdasarkan score', fontsize=16)
plt.xlabel('Score',fontsize=12)
plt.ylabel('Author',fontsize=12)
plt.xticks(rotation=45,ha='right')

# menampilkan plot
plt.show()
```

```python
# menggabungkan kolom 'Author' dan 'Book Name' untuk membuat label gabungan
df['Author_book']=df['Author'] + '-' + df['Book Name']

# ambil 10 buku teratas berdasarkan score terbanyak
top_10_books_by_score=df.sort_values(by='Score',ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.scatterplot(x='Score',y='Author_book',size='Score',hue='Score', sizes=(100,1000),
                data=top_10_books_by_score,palette='coolwarm',legend=False)

# menambahkan judul dan label
plt.title('Popularitas Author dan Judul Buku berdasarkan score', fontsize=16)
plt.xlabel('Score',fontsize=12)
plt.ylabel('Author-Judul Buku', fontsize=12)
plt.xticks(rotation=45,ha='right')

# menampilkan plot
plt.show()
```
